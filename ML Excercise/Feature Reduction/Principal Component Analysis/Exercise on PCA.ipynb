```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
encoder = LabelEncoder()

df.Sex = encoder.fit_transform(df.Sex)
df.ChestPainType = encoder.fit_transform(df.ChestPainType)
df.RestingECG = encoder.fit_transform(df.RestingECG)
df.ExerciseAngina = encoder.fit_transform(df.ExerciseAngina)
df.ST_Slope = encoder.fit_transform(df.ST_Slope)

df.head()
```

```python
def outliers(df, cols):
    plt.figure(figsize=(16,9))
    for i in range(0,4):
        for j in range(0,4):
            plt.subplot(4,4, i * 4 + j + 1)
            plt.boxplot(df[cols[i * 4 + j]],vert=False)
            if i * 4 + j + 1 == len(cols):
                break
        if i * 4 + j + 1== len(cols):
            break
    plt.show()
```

```python
outliers(df, df.columns)
```
