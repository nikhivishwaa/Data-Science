### MatPlot Lib charting

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.plot([1,5,9,11,14,15],[7,11,4,5,9,8])
plt.show()
```

### Plot two graph

```python
from matplotlib import style

style.use('ggplot')  # styling the plots

x = [5,10,12,14,15]
y = [5,10,15,20,25]
x1 = [4,7,12,14,17]
y1 = [5,10,15,20,25]

plt.plot(x,y)
plt.plot(x1,y1)

plt.show()
```

```python
x = [5,10,12,14,15]
y = [5,10,15,20,25]
x1 = [4,7,12,15,17]
y1 = [5,10,15,20,25]

plt.plot(x,y, label='first')
plt.plot(x1,y1, label='second')

# to show labels of both lines we have to legend f()

plt.legend()

plt.show()
```

```python
plt.plot(x,y, label='first', linewidth=6)
plt.plot(x1,y1, label='second')
plt.legend()

plt.show()
```

```python
plt.hist(x,y, label='first', linewidth=6)
plt.hist(x1,y1, label='second')
plt.legend()

plt.show()
```
